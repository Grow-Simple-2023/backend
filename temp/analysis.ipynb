```python
import json


with open("final.json", "r") as f:
    final = json.load(f)

with open("initial_info.json", "r") as f:
    initial_info = json.load(f)
    
with open("complete_info.json", "r") as f:
    complete_info = json.load(f)

with open("rider.json", "r") as f:
    rider = json.load(f)
    

with open("final.json", "r") as f:
    final = json.load(f)
```

```python
complete_info = complete_info['unassigned_items']
rider = rider['unassigned_riders']
```

```python
def item_no_to_lat_long(item_no):
    if item_no == -1: return (30, 100)
    else: return (complete_info[item_no]['location']['latitude'], complete_info[item_no]['location']['longitude'])
```

```python
l = []
for x in complete_info:
    l.append(x['id'])
    
r = []
for x in rider:
    r.append(x['phone_no'])

initial = {
    "item_ids": l,
    "rider_phone_nos": r
}
with open("initial_info.json", "w+") as f:
    json.dump(initial, f, indent = 3)
```

```python
import matplotlib.pyplot as plt

for route in final['routes']:
    route.append(route[0])

for route in final['routes']:
    plt.plot([item_no_to_lat_long(item_no)[0] for item_no in route], [item_no_to_lat_long(item_no)[1] for item_no in route], 'o-')
```
