```python
import json
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def tsp_plot(lat_longs):
    plt.plot(lat_longs[1:-1, 1], lat_longs[1:-1, 0], '-o', linewidth=2)
```

```python
with open("final.json", "r") as f:
    final = json.load(f)

for route in final["routes"]:
    
    lat_longs = []
    for item in route["items_in_order"]:
        lat_longs.append((item["location"]["latitude"], item["location"]["longitude"]))
    lat_longs = np.array(lat_longs)
    tsp_plot(lat_longs)
    
    
```
